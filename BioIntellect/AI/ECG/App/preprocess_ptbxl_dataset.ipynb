```python
from pathlib import Path

BASE = Path(r"D:\AOU-Graduation-Project\BioIntellect\AI\ECG\Data\ptb-xl-a-large-publicly-available-electrocardiography-dataset-1.0.3")

df["filename_hr"] = df["filename_hr"].apply(lambda p: str(BASE / Path(p)))
```

```python
df.drop(columns=["patient_id", "strat_fold"], inplace=True)
```

```python
train_folds = [1,2,3,4,5,6,7,8]
val_folds   = [9]
test_folds  = [10]

train_df = df[df["strat_fold"].isin(train_folds)].copy()
val_df   = df[df["strat_fold"].isin(val_folds)].copy()
test_df  = df[df["strat_fold"].isin(test_folds)].copy()

print("train:", train_df.shape, "val:", val_df.shape, "test:", test_df.shape)
```

```python
train_df.to_csv("ptbxl_train.csv")
val_df.to_csv("ptbxl_val.csv")
test_df.to_csv("ptbxl_test.csv")
```

```python
train_patients = set(train_df["patient_id"].dropna().astype(int))
val_patients   = set(val_df["patient_id"].dropna().astype(int))
test_patients  = set(test_df["patient_id"].dropna().astype(int))

print("train ∩ val:", len(train_patients & val_patients))
print("train ∩ test:", len(train_patients & test_patients))
print("val ∩ test:", len(val_patients & test_patients))
```
